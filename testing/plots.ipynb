```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv('merged.csv')
```

```python
counter = Counter(data['Rating'])
counter
d = {'Positive': [counter[1]], 'Negative': [counter[0]]}
barData = pd.DataFrame(d)
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
# Draw a nested barplot by species and sex
g = sns.barplot(data=barData, order=["Positive", "Negative"])
```
